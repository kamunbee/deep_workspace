```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터
x_data = [2,4,6,8,10,12,14]
y_data = [0,0,0,1,1,1,1]

print(x_data)
print(y_data)
```

```python
# 그래프로 나타내기
plt.scatter(x_data, y_data)
plt.show()
```

```python
# a와 b의 값
a = 0
b = 0
```

```python
# 학습률
lr = 0.05
```

```python
# 시그모이드 함수
def sigmoid(x) :
    return 1 / (1 + np.e ** (-x))
```

```python
# 반복 횟수
epochs = 20001
```

```python
# ndarray 생성
x_data2 = np.array(x_data)
y_data2 = np.array(y_data)

print(x_data2)
print(y_data2)
```

```python
for i in range(epochs) :
    # 예측 결과를 가져온다.
    y_pred = sigmoid(a * x_data2 + b)
    #print(y_pred)

    # 오차를 계산한다.
    error = y_pred - y_data2
#     print(error)
    
    # a 와 b에 대한 편미분을 수행한다.
    a_diff = ( 1/ len(x_data2)) * sum(x_data2 * error)
    b_diff = (1 / len(x_data2)) * sum(error)
    
    #print(a_diff)
    #print(b_diff)
    
    # a와 b의 값을 계산한다.
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    #print(a)
    #print(b)
    
    if i % 1000 == 0:
        print(f'epoch {i}, a : {a}, b : {b}')
        
        plt.scatter(x_data2, y_data2)
        # 그래프로 나타낼 x값의 범위
        x_range = (np.arange(0, 15, 0.1))
        plt.plot(x_range, np.array([sigmoid(a * x + b) for x in x_range]))
        plt.show()
```
