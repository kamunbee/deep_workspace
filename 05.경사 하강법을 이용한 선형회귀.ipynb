```python
# 공부 시간 x와 성적 y의 리스트를 만들기
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

print(x)
print(y)
```

```python
# 그래프로 그리기
plt.scatter(x, y)
plt.ylabel('성적')
plt.xlabel('학습시간')
plt.show()
```

```python
# 데이터를 ndarray로 생성한다.
x_data = np.array(x)
y_data = np.array(y)

print(x_data)
print(y_data)
```

```python
# 임의의 기울기와 절편
a = 0
b = 0
```

```python
# 학습률 정하기
# lr = 0.05
lr = 0.03
```

```python
# 몇번 반복할지..
epochs = 2001
```

```python
# 경사하강법
# 에포크 수 만큼 반복
for i in range(epochs) :
    # 현재의 기울기와 절편을 이용해 값을 예측
    y_pred = a * x_data + b
#     print(y_pred)
    
    # 오차를 구한다.
    error = y_data - y_pred
#     print(error)
    
    # 오차함수를 a로 미분한 값
    a_diff = -(2 / len(x_data)) * sum(x_data * error)
    # 오차함수를 b로 미분한 값
    b_diff = -(2 / len(x_data)) *sum(error)
    
    #print(a_diff)
    #print(b_diff)
    
    # 학습률을 곱해 기존의 a 값을 업데이트 한다.
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    # print(a)
    # print(b)
    
    # 100회 마다 기울기와 절편을 출력한다
    if i % 100 == 0:
        print(f'epoch : {i}, 기울기 : :{a}, 절편 : {b}')
```

```python
# 발견한 기울기와 절편을 통해 예측 결과를 추출한다.
y_pred =a * x_data + b
y_pred
```

```python
plt.scatter(x, y)
plt.scatter(x, y_pred)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)], "R")
plt.show()
```
